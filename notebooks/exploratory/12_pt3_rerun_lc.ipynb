### OY Population with No Diploma:

Previously, we missed data because we were searching for '1' instead of '01' for single-digit values:

```python
# Previously we missed data because we were searching of '1' instead of '01' for single-digit values:
skc_df[skc_df['education_attained'] == '1']
```

```python
# vs. :
skc_df[skc_df['education_attained'] == '01']
```

So we created a pad function to cast the single digit entries to include a '0' before them:

```python
# Create function to add 0's to single-digit entries:
def pad(num):
    if num < 10:
        return f'0{num}'
    else: 
        return str(num)  
```

```python
# What is the total number of 16-24 OY with no diploma?  Valid entries:  bb-15.
no_dip = ['bb'] + list(map(pad, range(1, 16)))
```

```python
no_dip_oy_df = oy_df[oy_df['education_attained'].isin(no_dip)]
no_dip_oy_df
```

```python
# Check all education_attained values are valid:
no_dip_oy_df['education_attained'].unique()
```

```python
# Now, calculate total OY with no diploma:
no_dip_oy_df['person_weight'].sum()
```

Breaking this down into the 3 age groups:

```python
dc.age_breakdown(no_dip_oy_df)
```
